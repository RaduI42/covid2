```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.impute import SimpleImputer
```

```python
covid_df =pd.read_csv('owid-covid-data.csv')
```

```python
covid_df.drop(["new_deaths_smoothed", "new_cases_per_million", "total_cases_per_million"], inplace=True, axis = 1)
```

```python
covid_df.head()
```

```python
covid_df.rename(columns={'date': 'Date','location':'Country','continent': 'Continent','iso_code':'ISO_code'},inplace=True)
```

```python
covid_df
```

```python
continent_unique = list(covid_df.Continent.unique())
```

```python
continent_unique
```

```python
imputer = SimpleImputer(strategy='constant')
```

```python
covid_df2 = pd.DataFrame(imputer.fit_transform(covid_df),columns=covid_df.columns)
```

```python
covid_df3 = covid_df2.groupby(['Date','Country',])[['total_cases','total_deaths','total_vaccinations']].sum().reset_index()
covid_df3
```

```python
covid_df3['total_cases'].replace({'missing_value': 0}, inplace=True)
```

```python
covid_df3['total_deaths'].replace({'missing_value': 0}, inplace=True)
covid_df3['total_vaccinations'].replace({'missing_value': 0}, inplace=True)
covid_df3
```

```python
covid_df4 = covid_df3[covid_df3['total_deaths']>100000]
```

```python
countries = covid_df4['Country'].unique()
len(countries)
```

```python
country_deaths_greaterthan1000000 = list(covid_df4.Country.unique())
```

```python
country_deaths_greaterthan1000000
```

```python
for idx in range(0,len(countries)):
    c = covid_df[covid_df['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(c)),c['total_cases'],color="blue",label="total_cases")
    plt.scatter(np.arange(0,len(c)),c['total_deaths'],color="red",label="total_deaths")
    plt.scatter(np.arange(0,len(c)),c['total_vaccinations'],color="green",label="total_vacciantions")
    plt.title(countries(idx))
    plt.xlable("number of days since first suspect")
    plt.ylable("number of cases")
    plt.legend()
    plt.show()
```
